```python
def ilinechart(df, x, y, groups=None, title=""):
    fig = px.line(df, x=x, y=y, color=groups, title=title, template="none").update(
        layout=dict(title=dict(x=0.5))
    )

    fig.show()
```

```python
daily_btc = pd.read_csv("BTC-USD.csv")
daily_spy = pd.read_csv("SPY.csv")
```

```python
daily_btc["Date"] = pd.to_datetime(daily_btc["Date"])
daily_spy["Date"] = pd.to_datetime(daily_spy["Date"])
```

```python
daily_btc.head()
```

```python
daily_spy.head()
```

```python
daily_btc["% Change"] = (
    (daily_btc["Close"] - daily_btc["Close"][0]) / daily_btc["Close"][0] * 100
)

daily_spy["% Change"] = (
    (daily_spy["Close"] - daily_spy["Close"][0]) / daily_spy["Close"][0] * 100
)

daily_btc["Change"] = (
    (daily_btc["Close"] - daily_btc["Close"].shift(1)) / daily_btc["Close"] * 100
)

daily_spy["Change"] = (
    (daily_spy["Close"] - daily_spy["Close"].shift(1)) / daily_spy["Close"] * 100
)
```

```python
daily_corr = daily_btc["Change"].corr(daily_spy["Change"])
daily_corr
```

```python
all_instruments = daily_spy.merge(daily_btc, left_on="Date", right_on="Date")
all_instruments = all_instruments.rename(
    columns={
        "% Change_x": "SPY % Change",
        "% Change_y": "BTC % Change",
        "Change_x": "SPY Daily % Change",
        "Change_y": "BTC Daily % Change",
    }
)

all_instruments = all_instruments[1:]
all_instruments
```

```python
all_instruments_melted = pd.melt(
    all_instruments,
    id_vars="Date",
    value_vars=["SPY Daily % Change", "BTC Daily % Change"],
)
all_instruments_melted
```

```python
ilinechart(
    all_instruments_melted, "Date", "value", groups="variable", title="Daily % Change"
)
```

```python
daily_btc["10 day"] = daily_btc["Close"].rolling(window=10).mean()
daily_btc["20 day"] = daily_btc["Close"].rolling(window=20).mean()
daily_btc["50 day"] = daily_btc["Close"].rolling(window=50).mean()
daily_btc["100 day"] = daily_btc["Close"].rolling(window=100).mean()
```

```python
daily_btc_melted = pd.melt(
    daily_btc,
    id_vars="Date",
    value_vars=["Close", "10 day", "20 day", "50 day", "100 day"],
)
```

```python
ilinechart(daily_btc_melted, "Date", "value", groups="variable", title="Daily BTC")
```
